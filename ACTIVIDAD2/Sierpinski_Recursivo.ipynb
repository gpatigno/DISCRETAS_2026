```python
def dibujar_triangulo(ax, puntos, color, etiqueta):
    x = [p[0] for p in puntos] + [puntos[0][0]]
    y = [p[1] for p in puntos] + [puntos[0][1]]
    
    ax.fill(x, y, color=color, edgecolor="black")
    
    # Centroide para colocar el número
    cx = sum(p[0] for p in puntos) / 3
    cy = sum(p[1] for p in puntos) / 3
    
    ax.text(cx, cy, str(etiqueta), 
            ha='center', va='center', 
            fontsize=8, color='white')
```
