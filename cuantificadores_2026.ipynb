```python
def cuantificador_universal(Dominio, P):
    for x in Dominio:
        if not P(x):
            return False
    return True
```

```python
def cuantificador_existencial(Dominio, P):
    for x in Dominio:
        if P(x):
            return True
    return False
```

```python
A = [1, -2, 3, -4]
B = [1, 2, 3, 4]
```

```python
resultadoA_Univ = cuantificador_universal(A, lambda x: x > 0)
print(resultadoA_Univ)
```

```python
resultadoB_Univ = cuantificador_universal(B, lambda x: x > 0)
print(resultadoB_Univ)
```

```python
resultadoA_Ex = cuantificador_existencial(A, lambda x: x < 0)
print(resultadoA_Ex)
```

```python
resultadoB_Ex = cuantificador_existencial(B, lambda x: x < 0)
print(resultadoB_Ex)
```
